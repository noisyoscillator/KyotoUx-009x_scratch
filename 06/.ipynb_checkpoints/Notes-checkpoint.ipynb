# Non-Trivial random walk

- First return time of ideal random walk, from Ref[5], IV.C, pg. 15.

    The idea is to find the probability that a random walker returns to zero for the first time, at time t.
    It is interesting because it yields a power law distribution with exponent $\alpha=-3/2$, which means that the mean diverges! 
    The average time to the first visit is infinite, eventhough the "typical" exit time is finite.
    The downside is it can be solved analytically.
    It can also be applied to describe the distribution of the lifetime of biological taxa (branches in the evolutionary tree of a given species).
    We can ask the students to derive this from the fossil data maybe?
    
    
- Gamblers ruin

    Gambling game between two players, A and B. Consider different gambling strategies depending on the probabilities of winning/losing a bet and money that is bet
    https://www.math.unl.edu/~sdunbar1/MathematicalFinance/Lessons/FirstStepAnalysis/RuinProbs/ruinprobs.pdf
    
    
- Exit probability or Exit time of a diffusing particle (related to previous example), from Ref[1], pg. 26-30.

    Check the probability that a particle will hit some specified boundary, this depends on the dimension of the problem.
