```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

# Model 1

$$\begin{align}
p_i(t + \Delta t) &= p_i(t) + c f_i(t), \qquad i=1,2\\
f_i(t) &= \begin{cases}
    +\Delta p & \mathrm{prob.} 1/2 \\
    -\Delta p & \mathrm{prob.} 1/2
\end{cases}
\end{align}$$


```python
def model1(params,p0,numt):
    mktprice = np.zeros(numt)
    ticktime = np.zeros(numt,dtype=np.int)
    price    = np.array([p0[0], p0[1]])
    time,tick = 0,0
    cdp      = params['c']*params['dp']
    while tick < numt:
        while np.abs(price[0]-price[1]) < params['L']:
            price = price + np.random.choice([-cdp,cdp], size=2)
            time += 1
        price[:] = np.average(price)
        mktprice[tick] = price[0]
        ticktime[tick] = time
        tick += 1
    return ticktime,mktprice
```

```python
np.random.seed(0)
params = {'L' : 0.01, 'c': 0.01, 'dp' : 0.01, 'dt' : 0.01**2}
```

```python
ticktime,mktprice = model1(params, [100.25, 100.25], 5000)
np.savetxt('model1.txt',np.transpose([ticktime, mktprice]))
```

```python
ticktime,mktprice = np.loadtxt('model1.txt', unpack=True)
timeinterval = (ticktime[1:] - ticktime[0:-1])*params['dt']
dprice       = mktprice[1:] - mktprice[0:-1]
```

```python
%matplotlib inline
fig,[ax,bx,cx] = plt.subplots(figsize=(18,6), ncols=3)

ax.plot(mktprice)
ax.set_ylim(99,101)
ax.set_xlabel(r'Time (ticks)')
ax.set_ylabel(r'Market price')

bx.plot(dprice)
bx.set_xlabel(r'Time (ticks)')
bx.set_ylabel(r'Price Change')

cx.plot(timeinterval)
cx.set_xlabel(r'Time (ticks)')
cx.set_ylabel(r'Transaction interval')
fig.tight_layout()
plt.show()
```

```python
# lets analyze the price change
avg,sig = np.average(dprice), np.var(dprice)
print(avg, sig)
gaussian= np.random.randn(5000)*np.sqrt(sig) + avg

fig, [ax,bx] = plt.subplots(figsize=(15,7.5),ncols=2)

ax.plot(dprice,label=r'Model')
ax.plot(gaussian,label=r'Gaussian',alpha=0.8)
for m in [1,2,3,4]:
    dy = m*np.sqrt(sig)
    ax.plot([0, 5000], np.ones(2)*(avg+dy), color='dimgray', ls='--',lw=2)
    ax.plot([0, 5000], np.ones(2)*(avg-dy), color='dimgray', ls='--',lw=2)
ax.set_xlabel(r'Time (ticks)')
ax.set_ylabel(r'Price Change')
ax.legend(fontsize=20)

bx.hist(dprice, bins=50, lw=0, normed=True)
bx.hist(gaussian, bins=50, lw=0, alpha=0.8, normed=True)
bx.set_xlabel(r'Price Change')
bx.set_ylabel(r'Probability')
plt.show()
```

## Notes
- The price change $dP(n) = P(n+1) - P(n)$ is clearly not Gaussian. 
- Notice that very large fluctuations, which would be impossible for a Gaussian process appear: we see points that are more than $4\sigma$ from the center!

```python
def pdf(data,bins=50):
    hist, edges = np.histogram(data, bins=bins, density=True)
    edges   = (edges[:-1] + edges[1:])/2.0
    nonzero = hist > 0.0
    return edges[nonzero], hist[nonzero]

fig, [ax,bx] = plt.subplots(figsize=(15,7.5), ncols=2)

# probability density of price change between ticks
dprice     = mktprice[1:] - mktprice[0:-1]
edges,hist = pdf(np.abs(dprice),bins=25)
ax.plot(edges, hist)
ax.set_xlabel(r'Price change',fontsize=20)
ax.set_ylabel(r'Probability Density',fontsize=20)
ax.set_xlim(0.0, 2.0e-2)
ax.set_ylim(1, 500)
ax.semilogy()

# probability density of transaction tim intervals between ticks
edges,hist = pdf(timeinterval, bins=100)
bx.plot(edges,hist)
bx.set_xlabel(r'Transaction Interval')
bx.update_datalim([[0.0, 2.0], [1e-2,2]])
bx.set_xlim([0.0, 2.0])
bx.set_ylim([1e-2,2])
bx.semilogy()
plt.show()
```

## Notes

- Tail of both distributions is characterized by an exponential
- For transaction interval, this means we have a Poisson process

# Model 3

$$\begin{align}
p_i(t + \Delta t) &= p_i(t) + d\langle\Delta P\rangle_M \Delta t + c f_i(t), \qquad i=1,2\\
f_i(t) &= \begin{cases}
    +\Delta p & \mathrm{prob.} 1/2 \\
    -\Delta p & \mathrm{prob.} 1/2
\end{cases}
\end{align}$$

$$
\langle\Delta P\rangle_M = \frac{2}{M (M+1)}\sum_{k=0}^{M-1}(M-k)\Delta P(n-k)
$$

```python
def model2(params,p0,numt):
    def avgprice(pn):
        M = len(pn)
        weights  = np.array(range(1,M+1))*2.0/(M*(M+1))
        return weights.dot(pn)
    
    mktprice = np.zeros(numt)
    dmktprice= np.zeros(numt)
    ticktime = np.zeros(numt,dtype=np.int)
    price    = np.array([p0[0], p0[1]])
    time,tick= 0,0
    deltapm  = 0.0
    cdp      = params['c']*params['dp']
    ddt      = params['d']*params['dt']
    while tick < numt:
        while np.abs(price[0]-price[1]) < params['L']:
            price = price + deltapm + np.random.choice([-cdp,cdp], size=2)
            time += 1
        price[:]       = np.average(price)
        mktprice[tick] = price[0]
        dmktprice[tick]= mktprice[tick] - mktprice[np.max([0,tick-1])]
        ticktime[tick] = time
        tick += 1
        tick0 = np.max([0, tick - params['M']])
        deltapm = avgprice(dmktprice[tick0:tick])*ddt
    return ticktime,mktprice,dmktprice
```

```python
np.random.seed(0)
params2 = {'L' : 0.01, 'c': 0.01, 'dp' : 0.01, 'dt' : 0.01**2, 'd':1.25, 'M':1}
```

```python
ticktime2,mktprice2,dprice2 = model2(params, [100.25, 100.25], 5000)
np.savetxt('model2.txt',np.transpose([ticktime2, mktprice2,dprice2]))
```

```python
ticktime2,mktprice2,dprice2 = np.loadtxt('model2.txt', unpack=True)
timeinterval2 = (ticktime2[1:] - ticktime2[0:-1])*params2['dt']
```

```python
%matplotlib inline
fig,[ax,bx,cx] = plt.subplots(figsize=(18,6), ncols=3)

ax.plot(mktprice2,label=r'model 2')
ax.plot(mktprice,alpha=0.8,label=r'model 1')
ax.set_xlabel(r'Time (ticks)')
ax.set_ylabel(r'Market price')
ax.legend(loc=2,fontsize=20)

bx.plot(dprice2)
bx.plot(dprice,alpha=0.8)
bx.set_xlabel(r'Time (ticks)')
bx.set_ylabel(r'Price Change')

cx.plot(timeinterval2)
cx.plot(timeinterval,alpha=0.8)
cx.set_xlabel(r'Time (ticks)')
cx.set_ylabel(r'Transaction interval')
fig.tight_layout()
plt.show()
```

```python
def pdf(data,bins=50):
    hist, edges = np.histogram(data, bins=bins, density=True)
    edges   = (edges[:-1] + edges[1:])/2.0
    nonzero = hist > 0.0
    return edges[nonzero], hist[nonzero]

fig, ax = plt.subplots(figsize=(7.5,7.5))
# probability density of price change between ticks
for data,lbl,mk,nb in zip([dprice2, dprice], [r'model2', r'model1'], ['o','s'], [50, 20]):
    edges,hist = pdf(np.abs(data),bins=nb)
    ax.plot(edges, hist, label=lbl)
    
dmyx = np.logspace(-2,-1)
ax.plot(dmyx,1e-4/dmyx**3, color='grey',ls='--', lw=2)

ax.legend(fontsize=20)
ax.set_xlabel(r'Price change',fontsize=20)
ax.set_ylabel(r'Probability Density',fontsize=20)
ax.set_xlim(1e-3, 0.5)
ax.set_ylim(5e-2, 2e2)
ax.semilogy()
ax.semilogx()
plt.show()
```
